```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
d = np.load("traces.npy")
print(d.shape)
plt.plot(d[0])
```

```python
nums = []
for x in d:
    s = np.argmax(x< -0.4)
    k = np.argmax(x[::-1] < -0.4)
    l = (len(x)-k)-s
    nums += [l]
plt.plot(nums)
```

```python
removed = []
for i,x in enumerate(nums):
    if x < 2100:
        removed.append(str(i))
print(f"flag{{{','.join(removed)}}}")
```
